```python
import numpy as np
import seaborn as sns
import matplotlib.colors as clr
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import squarify 
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.set(style="white", context="talk")
```

```python
genres = pd.read_csv('ab-2017-01-23-highlevel-genres.csv')
```

```python
genre_dortmund = genres['genre_dortmund']
gd_counts = pd.Series.value_counts(genre_dortmund)
```

```python
genre_rosamerica = genres['genre_rosamerica']
gr_counts = pd.Series.value_counts(genre_rosamerica)
```

```python
genre_tzanetakis = genres['genre_tzanetakis']
gt_counts = pd.Series.value_counts(genre_tzanetakis)
```

```python
fig, ax = plt.subplots(subplot_kw=dict(aspect="equal"))
wedges, texts = ax.pie(gd_counts)
pct = 100*gd_counts/len(genre_dortmund)

ax.legend(labels=['%s  %1.2f %%' % (l, s) for l, s in zip(list(pct.keys()), list(pct))], loc="center left",bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('genre_dortmund')
plt.show()
```

```python
total = float(len(genre_dortmund))
ax = sns.barplot(x=np.array(gd_counts), y=np.array(gd_counts.keys()), orient='h', palette='PuBu_r')

for p in ax.patches:
    loc = ax.patches[0].get_width()
    width = p.get_width()
    ax.text(loc+ 20, p.get_y()+p.get_height()/1.5, '  '+'{:1.2f}'.format(100*width/total) + '%', ha='left') 
        
ax.set_title('genre_dortmund')
ax.set_xticks([])
sns.despine(left=True, bottom=True)
```

```python
# Creating colorscheme
cmap = cm.Blues
mini=min(np.array(gd_counts))
maxi=max(np.array(gd_counts))
norm = clr.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in np.array(gd_counts)]

other = 0
x = np.array(gd_counts)
y = np.array(gd_counts.keys())

for el in x:
    
    pct = 100*el/float(len(genre_dortmund))
    
    if pct < 2:
        other += el
        i = np.where(x == el)
        x = np.delete(x, i, 0)
        y = np.delete(y, i, 0)
        
x = np.append(x, other)
y = np.append(y, 'other')

squarify.plot(sizes=x, label=y, alpha=.8, color=colors)
plt.axis('off')
plt.title('genre_dortmund')
plt.show() 
```

```python
fig, ax = plt.subplots(subplot_kw=dict(aspect="equal"))
wedges, texts = ax.pie(gr_counts)
pct = 100*gr_counts/len(genre_rosamerica)

ax.legend(labels=['%s  %1.2f %%' % (l, s) for l, s in zip(list(pct.keys()), list(pct))], loc="center left",bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('genre_rosamerica')
plt.show()
```

```python
total = float(len(genre_rosamerica))
ax = sns.barplot(x=np.array(gr_counts), y=np.array(gr_counts.keys()), orient='h', palette='PuBu_r')

for p in ax.patches:
    loc = ax.patches[0].get_width()
    width = p.get_width()
    ax.text(loc+ 20, p.get_y()+p.get_height()/1.5, '  '+'{:1.2f}'.format(100*width/total) + '%', ha='left') 
        
ax.set_title('genre_rosamerica')
ax.set_xticks([])
sns.despine(left=True, bottom=True)
```
