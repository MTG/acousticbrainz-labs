```python
import numpy as np
import pandas as pd
import graphs
```

```python
graphs.init_setup()
```

```python
genres = pd.read_csv('ab-2017-01-23-highlevel-genres.csv')
```

```python
genre_dortmund = genres['genre_dortmund']
gd_counts = pd.Series.value_counts(genre_dortmund)
```

```python
genre_rosamerica = genres['genre_rosamerica']
gr_counts = pd.Series.value_counts(genre_rosamerica)
```

```python
genre_tzanetakis = genres['genre_tzanetakis']
gt_counts = pd.Series.value_counts(genre_tzanetakis)
```

```python
graphs.genre_pie(gd_counts, genre_dortmund, 'genre_dortmund')
graphs.genre_pie(gr_counts, genre_rosamerica, 'genre_rosamerica')
graphs.genre_pie(gt_counts, genre_tzanetakis, 'genre_tzanetakis')
```

```python
graphs.genre_bar(gd_counts, genre_dortmund, 'genre_dortmund')
graphs.genre_bar(gr_counts, genre_rosamerica, 'genre_rosamerica')
graphs.genre_bar(gt_counts, genre_tzanetakis, 'genre_tzanetakis')
```

```python
graphs.genre_map(gd_counts, genre_dortmund, 'genre_dortmund', 2)
graphs.genre_map(gr_counts, genre_rosamerica, 'genre_rosamerica', 0)
graphs.genre_map(gt_counts, genre_tzanetakis, 'genre_tzanetakis', 1)
```
