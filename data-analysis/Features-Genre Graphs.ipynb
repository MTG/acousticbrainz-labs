```python
import numpy as np
import seaborn as sns
import matplotlib.colors as clr
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import squarify 
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.set(style="white", context="talk")
```

```python
genres = pd.read_csv('ab-2017-01-23-highlevel-genres.csv')
```

```python
genre_rosamerica = genres['genre_rosamerica']
g_mbids = genres['mbid']
g_dict = dict(zip(g_mbids, genre_rosamerica))
```

```python
features = pd.read_csv('ab-2017-01-23-loudness-features.csv')
```

```python
avg_loudness = features['average_loudness']
f_mbids = features['recording_mbid']
l_dict = dict(zip(f_mbids, avg_loudness))
```

```python
x = []
y = []
for mbid in g_dict:
    
    if mbid in l_dict: 
        x.append(g_dict[mbid])
        y.append(l_dict[mbid])          
```

```python
ax = sns.boxplot(x=np.array(x), y=np.array(y), palette='Set3')
sns.despine(left=True, bottom=True)
ax.set_title('genre_rosamerica')
```

```python
ax = sns.violinplot(x=np.array(x), y=np.array(y), palette="Set3", bw=.2, cut=1, lineswidth=1)
sns.despine(left=True, bottom=True)
ax.set_title('genre_rosamerica')
```
